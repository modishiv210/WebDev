```python
import requests
from bs4 import BeautifulSoup
```

```python
# Read the URL and get the HTML,
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'
}

# You will be scraping the data from this page
url = 'https://www.billboard.com/charts/hot-100'

# Use the requests library to get the HTML code at the url above
data = requests.get(url=url, headers=headers)

# The BeautifulSoup library makes it easy to parse HTML code
soup = BeautifulSoup(data.text, 'html.parser')
```

```python
# Using select
path = "#post-1479786 > div > div > div > div > div > div"
songs = soup.select(path)
```

```python
for i in range(1, len(songs)):
    rank_path = "ul > li > span"
    song_path = "ul > li > ul > li > h3"
    artist_path = "ul > li > ul > li > span"
    
    rank = songs[i].select_one(rank_path)
    song = songs[i].select_one(song_path)
    artist = songs[i].select_one(artist_path)

    if rank is not None: #and song is not None:
        rank = rank.text.strip()
        song = song.text.strip()
        artist = artist.text.strip()
        print(rank, '/', song, '/', artist)
```
